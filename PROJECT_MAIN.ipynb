```python
def count_outliers_iqr(series, factor=1.5):
    """Return number of outliers in a column using IQR."""
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    return ((series < lower_bound) | (series > upper_bound)).sum()

# Build summary table
outlier_summary = pd.DataFrame({
    "Column": numerical_column.columns,
    "Outlier_Count": [count_outliers_iqr(numerical_column[col]) for col in numerical_column.columns]
})

print(outlier_summary)
```

```python
def labeled_barplot(data, feature, perc=False, n=None):

    total = len(data[feature])
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))

    plt.xticks(rotation=90, fontsize=15)

    ax = sns.countplot(
        data=data,
        x=feature,
        hue=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n].sort_values(),
        legend=False
    )

    for p in ax.patches:
        if perc:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )
        else:
            label = p.get_height()

        x = p.get_x() + p.get_width() / 2
        y = p.get_height()

        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",)

    plt.show()
```

```python

# Choose two variables
x_col = "Charging_Type"                # categorical (X-axis)
y_col = "Charge_Time_hr"   # numerical (Y-axis)

# Group and aggregate (mean in this case)
data_grouped = data.groupby(x_col)[y_col].mean()

# Plot bar graph
plt.figure(figsize=(12,5))
plt.bar(data_grouped.index, data_grouped.values, color="skyblue", edgecolor="black")

plt.title(f"Average {y_col} by {x_col}")
plt.xlabel(x_col)
plt.ylabel(f"Average {y_col}")
plt.xticks(rotation=90)
plt.show()
```

```python
# Create Range Categories
data["Range_Category"] = pd.cut(
    data["Range_km"],
    bins=[0, 200, 300, 400, 500, 600],
    labels=["0-200 km", "201-300 km", "301-400 km", "401-500 km", "501-600 km"]
)

# Calculate Average Price by Range Category
avg_price_by_range = data.groupby("Range_Category", observed=False)["Price_USD"].mean()

# Plot Bar Chart
plt.figure(figsize=(8,5))
avg_price_by_range.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Comparison of Range v/s Price", fontsize=14)
plt.ylabel("Average Price (USD)", fontsize=12)
plt.xlabel("Range Category", fontsize=12)
plt.xticks(rotation=45)

# Add value labels on top of each bar
for i, value in enumerate(avg_price_by_range):
    plt.text(i, value, f"${value:,.0f}", ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#1 Which manufacturers dominate the EV market represented in the dataset?
make_counts = data['Manufacturer'].value_counts(normalize=True) * 100
make_market_share = make_counts.reset_index()
make_market_share.columns = ['Manufacturer', 'Market Share (%)']

plt.figure(figsize=(12,6))
sns.barplot(data=make_market_share.head(10), x='Manufacturer', y='Market Share (%)', hue='Manufacturer',palette="mako",legend=False)
plt.title("Top 10 EV Manufacturers by Market Share (%)")
plt.xticks(rotation=45)
plt.show()

```

```python
#2 What are the observable trends across registration years (adoption timeline)?
# Use the 'Year' column as registration year
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')

# Drop missing values just in case
data = data.dropna(subset=['Year'])

# Count number of registrations per year
yearly_counts = data['Year'].value_counts().sort_index()

# Plot adoption trend
plt.figure(figsize=(10,6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', linewidth=2)
plt.title("EV Adoption Timeline by Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Registrations")
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(yearly_counts.index, rotation=45)
plt.show()

# Print summary statistics
print("Adoption Trend Overview:")
print(yearly_counts)
```

```python
#3 Is there a correlation between MSRP and range/battery capacity?
subset = data[['Price_USD', 'Range_km', 'Battery_Capacity_kWh']].dropna()

# Correlation matrix
corr_matrix = subset.corr()
print("Correlation Matrix:")
print(corr_matrix)

# Heatmap for correlation
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Price, Range, and Battery Capacity")
plt.show()
```

```python
#4: How do different countries compare in EV manufacturing levels by total units sold in 2024?
from matplotlib.ticker import MultipleLocator, StrMethodFormatter
# Group by country and sum units sold
country_output = data.groupby("Country_of_Manufacture")["Units_Sold_2024"].sum().sort_values(ascending=False)

#  Select Top countries
top_countries = country_output.head(13)

# Plot bar chart
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar', color='royalblue', edgecolor='black')
plt.title("EV Sales by Country of Manufacturer - 2024", fontsize=14)
plt.xlabel("Country of Manufacture")
plt.ylabel("Units Sold (Count)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# y-axis spacing to 200,000
plt.gca().yaxis.set_major_locator(MultipleLocator(200000))

# y-axis numbers with commas
plt.gca().yaxis.set_major_formatter(StrMethodFormatter("{x:,.0f}"))

plt.show()

# Table of values
print("EV Sales by Country of Manufacturer - 2024):")
print(top_countries)
```
